# üìä Tabela-Resumo ‚Äî An√°lise da M√©dia de 2018 a 2023

Esta c√©lula descreve as tabelas e gr√°ficos que ser√£o gerados neste notebook, com base na m√©dia das vari√°veis do per√≠odo de 2018 a 2023.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Configura√ß√µes
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Caminho para o arquivo CSV
caminho = r'C:/Users/john-/OneDrive - Universidade Federal da Para√≠ba/√Årea de Trabalho/Artigo - Estat√≠stica/projeto-evasao/data/processed/ml_iq_2018_2023.csv'

# Leitura dos dados
df = pd.read_csv(caminho)

# Lista de vari√°veis quantitativas
variaveis = [
    'NOTA_PADRONIZADA_-_IDD',
    'NOTA_PADRONIZADA_-_ORGANIZA√á√ÉO_DID√ÅTICO-PEDAG√ìGICA',
    'NOTA_PADRONIZADA_-_INFRAESTRUTURA_E_INSTALA√á√ïES_F√çSICAS',
    'NOTA_PADRONIZADA_-_OPORTUNIDADE_DE_AMPLIA√á√ÉO_DA_FORMA√á√ÉO',
    'NOTA_PADRONIZADA_-_MESTRES',
    'NOTA_PADRONIZADA_-_DOUTORES',
    'MEDIA_CONCEITO_ENADE_(CONT√çNUO)',
    'IDD_(CONT√çNUO)',
    'CONCEITO_M√âDIO_DE_GRADUA√á√ÉO',
    'IGC_(CONT√çNUO)',
    'SALARIO',
    'QT_VG_TOTAL_MEDIA'
]
```

## üìå Estat√≠sticas Descritivas (Geral)

```python
df_media = df[variaveis].mean().to_frame(name="M√©dia")
df_media["Mediana"] = df[variaveis].median()
df_media["Desvio Padr√£o"] = df[variaveis].std()
df_media["M√≠nimo"] = df[variaveis].min()
df_media["M√°ximo"] = df[variaveis].max()
df_media["Coef. Varia√ß√£o (%)"] = (df_media["Desvio Padr√£o"] / df_media["M√©dia"]) * 100
df_media
```

## üî• Heatmap de Correla√ß√£o

```python
correlacoes = df[variaveis].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlacoes, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap de Correla√ß√£o das Vari√°veis")
plt.show()
```

## üì¶ Boxplots

```python
for var in variaveis:
    sns.boxplot(x=df[var])
    plt.title(f"Boxplot: {var}")
    plt.xlabel(var)
    plt.show()
```

## üìà Histogramas + Densidade

```python
for var in variaveis:
    sns.histplot(df[var], kde=True, bins=30)
    plt.title(f"Histograma + Densidade: {var}")
    plt.xlabel(var)
    plt.ylabel("Frequ√™ncia")
    plt.show()
```

## üìä Gr√°fico de Barras das M√©dias

```python
df_media_sorted = df_media.sort_values("M√©dia", ascending=False)
df_media_sorted["M√©dia"].plot(kind='barh', color='skyblue')
plt.title("M√©dia das Vari√°veis (2018‚Äì2023)")
plt.xlabel("Valor M√©dio")
plt.grid(True)
plt.show()
```

## üîÑ Gr√°ficos de Dispers√£o com Linha de Tend√™ncia

```python
pares = [
    ("IGC_(CONT√çNUO)", "SALARIO"),
    ("MEDIA_CONCEITO_ENADE_(CONT√çNUO)", "NOTA_PADRONIZADA_-_MESTRES"),
    ("NOTA_PADRONIZADA_-_IDD", "IDD_(CONT√çNUO)"),
]

for x, y in pares:
    sns.scatterplot(x=df[x], y=df[y])
    sns.regplot(x=df[x], y=df[y], scatter=False, color='red')
    plt.title(f"Dispers√£o: {x} vs {y}")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
```
